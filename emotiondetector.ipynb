```python
import requests
!pip install deepface
#Downloading real face images
!wget -O happy_face.jpg https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg
!wget -O sad_face.jpg https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg
#Importing required libraries
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
#Running emotion detection
result = DeepFace.analyze(img_path = "happy_face.jpg",
                          actions = ['emotion'],
                          enforce_detection=False)
#Printing the dominant emotion
print("Detected emotion:", result[0]['dominant_emotion'])
print("Confidence scores:", result[0]['emotion'])
#Showing the image with emotion label
img = cv2.imread("happy_face.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title(f"Emotion: {result[0]['dominant_emotion']}")
plt.axis('off')
plt.show()
```
